# Reynolds stress calculation

 $$
    \underbrace{\rho_0\mathbf{\hat{u}}_i\cdot\nabla\cdot(\overline{h_i\mathbf{u_i''}\otimes\mathbf{u_i''}})}_\text{Reynolds stress} = \rho_0 \left( \hat{u}_i\left(\overline{ \frac{\partial}{\partial x}(h_iu_i''^2) +\frac{\partial}{\partial y}(h_iu_i''v_i'')}\right) + \hat{v}_i\left(\overline{ \frac{\partial}{\partial x}(h_iu_i''v_i'') +\frac{\partial}{\partial y}(h_iv_i''^2)}\right)\right)
$$

$$ \frac{\partial A}{\partial x} = \frac{\partial A}{\partial lat}\frac{\partial lat}{\partial x} = \frac{\partial A}{\partial lat}\frac{\Delta lat}{dxt} $$ where dxt in metres

xarray diff does $\frac{\partial A}{\partial lat}$ as it does derivatives taking account of coordinate values, which are latitudes

delta lon is 0.1, delta lat depends on latitude (difference in consecutive latitudes)

Used 4 scripts, save_summed_Reynolds_termn.py where n is 1,2,3,4

E.g. save_summed_Reynolds_term1.py:

        h_binned = xr.concat([h_binned,h_binned_i], dim = 'time')
        vh_binned = xr.concat([vh_binned,vh_binned_i], dim = 'time')

    v = v.where(v!=np.inf).where(v !=-np.inf).fillna(0)
    u_primeprime = u-u_hat
    v_primeprime = v-v_hat

    huu = h_binned*u_primeprime**2

# Form stress calculation

$
    \underbrace{ \mathbf{\hat{u}}_i \cdot \overline{h_i'\nabla\phi_i'}}_\text{Form stress} = \frac{\overline{u_ih_i}}{\overline{h_i}}\overline{h_i'\frac{\partial}{\partial x} \phi_i'}+ \frac{\overline{v_ih_i}}{\overline{h_i}}\overline{h_i'\frac{\partial}{\partial y} \phi_i'}
$

from save_summed_Form_stress.py, which uses save_Montgomery.py

### save_Montgomery.py:
    """
    Calculates 1 month of Montgomery potential for full Southern Ocean
    """

    # Load modules

    # Standard modules
    import cosima_cookbook as cc
    import matplotlib.pyplot as plt
    import netCDF4 as nc
    import xarray as xr
    import numpy as np
    from dask.distributed import Client
    import cftime
    import glob
    import dask.array as dsa
    from cosima_cookbook import distributed as ccd
    # Ignore warnings
    import logging
    logging.captureWarnings(True)
    logging.getLogger('py.warnings').setLevel(logging.ERROR)


    if __name__ == '__main__':

    # Start a dask cluster with multiple cores
    client = Client(n_workers=8, local_directory='/scratch/x77/cy8964/dask_dump/dask_worker_space')
    # Load database
    session = cc.database.create_session('/g/data/ik11/databases/cosima_master.db')

    # This script just sums binned quantities for a single year.

    #### get run count argument that was passed to python script ####
    import sys
    run_count = int(sys.argv[1])
    month = str(run_count).zfill(2)
    year = str(sys.argv[2])

    expt = '01deg_jra55v13_ryf9091'

    time_slice= year + '-' + month
    start_time = year + '-01-01'
    end_time= year + '-12-31'
    
    # reference density value:
    rho_0 = 1035.0
    g = 9.81
    # Note: change this range, so it matches the size of your contour arrays:
    ## FULL SO ##
    lat_range = slice(-70,-34.99)
    lat_range_big =  slice(-70.05,-34.90)
    # Load saved binned h for the given month of year
    saved_dir = '/g/data/x77/cy8964/Post_Process/New_SO/SO_saving_energy_terms_'
    chunky = {"time":1,"isopycnal_bins":5,"yt_ocean":218,"xt_ocean":400 }
    h_binned = xr.open_dataset(saved_dir+'h_binned_'+year+'-'+month+'_test.nc', chunks = chunky).h_binned

    # Load eta_t for given month of year
    eta_t = cc.querying.getvar(expt,'eta_t',session,ncfile='%daily%',start_time=start_time, end_time=end_time)
    eta_t = eta_t.sel(yt_ocean=lat_range, time=time_slice)
    
    # Load 10yr mean reference depths eta_ref
    eta_ref = xr.open_dataset('/g/data/x77/cy8964/Post_Process/New_SO/eta_ref_10yr.nc', 
                           chunks = {'isopycnal;_bins':5,'yt_ocean':218,'xt_ocean':400}).eta_ref
    
    ## define isopycnal bins   
    isopycnal_bins_sigma1 = 1000+ np.array([1,28,29,30,31,31.5,31.9,32,32.1,32.2,32.25,32.3,32.35,32.4,32.42,32.44,32.46,32.48,32.50,32.51,
                            32.52,32.53,32.54,32.55,32.56,32.58,32.6,32.8,33,34,45])

    ## since bins are edges of bins, define middle densities to be used in 
    middle_density = (isopycnal_bins_sigma1+ np.roll(isopycnal_bins_sigma1, -1))/2
    middle_density[0]=1028.5
    middle_density[-2]=1034.5
    middle_density[-1]=1035

    ## define g primes - only make sense from 2nd entry
    g_prime = g*(middle_density - np.roll(middle_density, 1))/rho_0
    g_prime[0]=g
    g_prime = xr.DataArray(g_prime, coords = [isopycnal_bins_sigma1], dims = ['isopycnal_bins'], name = 'g_prime')

    
    #deviation of top of interface height from average
    etas = h_binned.roll(isopycnal_bins = 1,roll_coords = False ).sel(isopycnal_bins = slice(1020,1050)).cumsum('isopycnal_bins')
    etas_0 = xr.zeros_like(h_binned.isel(isopycnal_bins=0))
    etas = xr.concat([etas_0,etas], dim = 'isopycnal_bins')
    etas = etas+eta_t-eta_ref
    
    # load etas (uses a lot of memory)
    etas = ccd.compute_by_block(etas)
    
    # save and delete etas
    save_dir = '/g/data/x77/cy8964/Post_Process/New_SO/'
    ds = xr.Dataset({'etas': etas})
    ds.to_netcdf(save_dir+'etas_'+year+'-'+month+'.nc')

    del etas
    del h_binned
    del eta_t
    
    # load saved file
    etas = xr.open_dataset('/g/data/x77/cy8964/Post_Process/New_SO/etas_'+year+'-'+month+'.nc', chunks = {'time':1, 'isopycnal;_bins':5,'yt_ocean':218,'xt_ocean':400}).etas

    # compute MP anomaly
    phi = rho_0*g_prime*etas
    phi = phi.cumsum('isopycnal_bins')
    phi = ccd.compute_by_block(phi)
    
    # save MP anomaly
    save_dir = '/g/data/x77/cy8964/Post_Process/New_SO/'
    ds = xr.Dataset({'phi': phi})
    ds.to_netcdf(save_dir+'Montgomery_Potential_Anomaly_'+year+'-'+month+'.nc')

# EKE Budget

$$
    \partial_tEKE_i =- \nabla \cdot (\mathbf{\hat{u}_i}EKE_i) - \nabla \cdot (\overline{\mathbf{u}_i''EKE_i}) - \overline{\mathbf{u}_i''\cdot h_i\nabla \phi_i''} + \underbrace{\rho_0(\mathbf{\hat{u}}_i\cdot\nabla)\cdot(\overline{h_i\mathbf{u_i''}\otimes\mathbf{u_i''}})}_\text{Reynolds stress} +  \rho_0 \overline{h_i\mathbf{F}_{\tau,i}\cdot \mathbf{u}_i''}
$$


```python
v_hat = xr.open_dataset('/g/data/x77/cy8964/Post_Process/New_SO/v_hat_10yr_dzt.nc').v_hat
v_hat = v_hat.load()
u_hat = xr.open_dataset('/g/data/x77/cy8964/Post_Process/New_SO/u_hat_10yr_dzt.nc').u_hat
u_hat = u_hat.load()
h_bar = xr.open_dataset('/g/data/x77/cy8964/Post_Process/New_SO/h_bar_10yr_dzt.nc').h_bar
h_bar = h_bar.load()
```

# Budget term 2
$\nabla \cdot (\mathbf{\hat{u}_i''}EKE_i) = \partial_x (\overline{u''1/2h\rho_0(u''^2+v''^2)}) + \partial_y (\overline{v''1/2h\rho_0(u''^2+v''^2)})$

Save $\overline{v''1/2h\rho_0(u''^2+v''^2)}$ and $\overline{u''1/2h\rho_0(u''^2+v''^2)}$ summed over yearly, then take gradient as for term 1

load in 10 years and save mean

    """

    """

    # Start a dask cluster with multiple cores

    import sys
    year = str(sys.argv[1])
    expt = '01deg_jra55v13_ryf9091'

    start_time = year + '-01-01'
    end_time= year + '-12-31'

    # reference density value:
    rho_0 = 1035.0
    # Note: change this range, so it matches the size of your contour arrays:
    ## FULL SO ##
    lat_range = slice(-70,-34.99)
    lat_range_big =  slice(-70.05,-34.90)

    #load coordinates
    yt_ocean = cc.querying.getvar(expt,'yt_ocean',session,n=1)

    # load 1 year of data

    month = '01'

    vh_binned = xr.open_dataset(saved_dir+'vh_binned_'+year+'-'+month+'_test.nc', chunks = chunky).vh_binned
    uh_binned = xr.open_dataset(saved_dir+'uh_binned_'+year+'-'+month+'_test.nc', chunks = chunky).uh_binned

    for i in np.arange(2,13):
        month = str(i).zfill(2)
        h_binned_i = xr.open_dataset(saved_dir+'h_binned_'+year+'-'+month+'_test.nc', chunks = chunky)
        h_binned_i = h_binned_i.h_binned

        vh_binned_i = xr.open_dataset(saved_dir+'vh_binned_'+year+'-'+month+'_test.nc', chunks = chunky)

        uh_binned_i = uh_binned_i.uh_binned

        h_binned = xr.concat([h_binned,h_binned_i], dim = 'time')
        vh_binned = xr.concat([vh_binned,vh_binned_i], dim = 'time')
        uh_binned = xr.concat([uh_binned,uh_binned_i], dim = 'time')

    #load 10 year means of data
    chunky2 = {"isopycnal_bins":5,"yt_ocean":218,"xt_ocean":400 }
    saved_dir2 = '/g/data/x77/cy8964/Post_Process/New_SO/'
    h_bar = xr.open_dataset(saved_dir2+'h_bar_10yr.nc', chunks = chunky2).h_bar
    u_hat = xr.open_dataset(saved_dir2+'u_hat_10yr.nc', chunks = chunky2).u_hat
    v_hat = xr.open_dataset(saved_dir2+'v_hat_10yr.nc', chunks = chunky2).v_hat

    # define primed quantities
    u = uh_binned/h_binned/rho_0

    v = vh_binned/h_binned/rho_0
    v = v.where(v!=np.inf).where(v !=-np.inf).fillna(0)
    u_primeprime = u-u_hat
    v_primeprime = v-v_hat
    
    # define terms1 and 2
    term1 = rho_0/2*h_binned*u_primeprime*(u_primeprime**2+v_primeprime**2)
    term1 = term1.sum('time')

    term2 = term2.sum('time')

    ds.to_netcdf(save_dir+'EKE_budget_sum_term2_term1'+year+'.nc')    
    del term1
    
    #save term2
    term2 = ccd.compute_by_block(term2)
    
    save_dir = '/g/data/x77/cy8964/Post_Process/New_SO/'
    ds = xr.Dataset({'term2': term2})
    ds.to_netcdf(save_dir+'EKE_budget_sum_term2_term2'+year+'.nc')
