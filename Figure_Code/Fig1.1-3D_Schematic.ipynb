## Code to make schematic with EKE on surface and density on side

Adapted from code from Adele Morrison

```python
%matplotlib inline

import cosima_cookbook as cc
import matplotlib.pyplot as plt
import numpy as np
import netCDF4 as nc
import xarray as xr
import cmocean as cm
from collections import OrderedDict
import cftime
import glob
from gsw import SA_from_SP, p_from_z, sigma1
from cosima_cookbook import distributed as ccd

import logging
logging.captureWarnings(True)
logging.getLogger('py.warnings').setLevel(logging.ERROR)

from dask.distributed import Client
```

```python
session = cc.database.create_session()#'/g/data/ik11/databases/cosima_master.db')
```

```python
client = Client(n_workers=4)
client
```

```python
expt = '01deg_jra55v13_ryf9091'
pot_rho_2 = cc.querying.getvar(expt, 'pot_rho_2', session, start_time = '2170-01-01').sel(time = slice('2170-01-01','2171-01-01'))
```

```python
exp = '01deg_jra55v13_ryf9091'

latN =10
latS = -90
lat_slice = slice(latS,latN)
lonW = -220
lonE = 0
lon_slice = slice(lonW,lonE)
start_time = '2170-01-01'
end_time = '2170-12-31'

depth_slice = slice(0,2000)
```

```python
ht = cc.querying.getvar('01deg_jra55v13_ryf9091','ht',session,n=1)
ht = ht.sel(yt_ocean=lat_slice).sel(xt_ocean=lon_slice)

land_mask = np.squeeze(ht.values)*0
land_mask[np.isnan(land_mask)] = 1
land_mask = np.where(land_mask==1,land_mask,np.nan)

# mask random islands in Pacific:
#land_mask[220:,:900] = np.nan

ht = ht.fillna(0)

Xt_land,Yt_land = np.meshgrid(ht.xt_ocean,ht.yt_ocean)
```

```python
lat_array = pot_rho_2*0 + pot_rho_2.yt_ocean
lon_array = pot_rho_2*0 + pot_rho_2.xt_ocean


```

```python
density = pot_rho_2.isel(time = 0)
density.load()
EKE = xr.open_dataset('/g/data/x77/cy8964/Post_Process/eddy_energies_EKE_2086-01-01_2087-01-01.nc').EKE
```

```python
EKE = EKE.rename({'xu_ocean':'xt_ocean','yu_ocean':'yt_ocean'})
EKE = EKE.sel(xt_ocean =lon_slice,yt_ocean = lat_slice)

EKE.load()
```

```python
# density front snapshot:
rho0 = pot_rho_2.isel(time = 0).sel(xt_ocean =lon_slice,yt_ocean = lat_slice)
rho0_front = rho0.sel(xt_ocean=lonE,method='nearest')
# mask
# coords for plotting:
Yf,Zf = np.meshgrid(rho0_front.yt_ocean,rho0_front.st_ocean)

# density top snapshot:
rho0_top = 1035*EKE.sum('st_ocean').where(EKE.sum('st_ocean')>0).sel(xt_ocean =lon_slice,yt_ocean = lat_slice)
# coords for plotting:
Xt,Yt = np.meshgrid(rho0_top.xt_ocean,rho0_top.yt_ocean)
Zt = np.zeros_like(Yt)

# density side snapshot:
# shift slightly to avoid topographic bumps:
rho0_side = rho0.sel(yt_ocean=latN-.2,method='nearest')
# coords for plotting:
Xs,Zs = np.meshgrid(rho0_side.xt_ocean,rho0_side.st_ocean)
```

```python
# side topo:

# calculate topo on side face:
st_edges_ocean = cc.querying.getvar('01deg_jra55v13_ryf9091','st_edges_ocean',session,
                 start_time=start_time, end_time=end_time,n=1).sel(st_edges_ocean=depth_slice)
st_edges_array = st_edges_ocean.expand_dims({'xt_ocean':ht.xt_ocean},axis=[1])

htN = ht.sel(yt_ocean=latN,method='nearest')
st_edges_with_partial = st_edges_array.where(st_edges_array<htN,other=htN)

# mask where these are equal:
side_topo = np.where(st_edges_with_partial == st_edges_array,np.nan,st_edges_with_partial)
side_topo = side_topo[:-1,:]
side_topo[-1,:] = st_edges_array[-1,:]

# limit depth:
side_topo = side_topo[:Xs.shape[0],:]

# remove spikes:
side_topo[:,:1000] = np.nan

# set bottom row to max depth:
side_topo[-1,:] = Zs.max()
# 
Ys = np.zeros_like(Xs)+latN
```

```python
# front face topo:

htW = ht.sel(xt_ocean=lonW,method='nearest')
st_edges_array = st_edges_ocean.expand_dims({'yt_ocean':ht.yt_ocean},axis=[1])
st_edges_with_partial = st_edges_array.where(st_edges_array<htW,other=htW)

# mask where these are equal:
front_topo = np.where(st_edges_with_partial == st_edges_array,np.nan,st_edges_with_partial)
front_topo = front_topo[:-1,:]
front_topo[-1,:] = st_edges_array[-1,:]

# limit depth:
front_topo = front_topo[:Xs.shape[0],:]

# set bottom row to max depth:
front_topo[-1,:] = Zf.max()

# 
Xf = np.zeros_like(Yf)+lonW
```

```python
#Black sheet

rho0_sheet = rho0.sel(yt_ocean=-90,method='nearest')
# mask
# coords for plotting:
Xss,Zss = np.meshgrid(rho0_sheet.xt_ocean,rho0_sheet.st_ocean)
```

```python
import matplotlib.colors as col

```

```python
newcmap = cm.tools.crop_by_percent(cm.cm.dense, 25, which='max', N= None)
```

```python
dset = xr.merge([rho0_top])
ds = dset
mask_ocean = 3 * np.ones((ds.dims['yt_ocean'], ds.dims['xt_ocean'])) * np.isfinite(ds.EKE)  
mask_land = 1 * np.ones((ds.dims['yt_ocean'], ds.dims['xt_ocean'])) * np.isnan(ds.EKE)  
mask_array = mask_ocean + mask_land
mask_array.plot()
```

```python
dset = xr.merge([rho0_front])
ds = dset
mask_ocean = 3 * np.ones((ds.dims['st_ocean'], ds.dims['yt_ocean'])) * np.isfinite(ds.pot_rho_2)  
mask_land = 1 * np.ones((ds.dims['st_ocean'], ds.dims['yt_ocean'])) * np.isnan(ds.pot_rho_2)  
mask_array2 = mask_ocean + mask_land
mask_array2.plot()
```

```python
dset = xr.merge([rho0_sheet])
ds = dset
mask_ocean = 3 * np.ones((ds.dims['st_ocean'], ds.dims['xt_ocean'])) * np.isfinite(ds.pot_rho_2)  
mask_land = 1 * np.ones((ds.dims['st_ocean'], ds.dims['xt_ocean'])) * np.isnan(ds.pot_rho_2)  
mask_array3 = mask_ocean + mask_land
mask_array3.plot()
```

```python
plt.rc('text', usetex=True)
plt.rcParams.update({'font.size': 18})
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d',box_aspect=(4.5,9,4))
stride_value = 1

clevs2 = np.append(np.linspace(0,0.16,9),1)
norm = col.Normalize(vmin=0, vmax=0.16)

cset = ax.contourf(Xt, Yt,rho0_top.values/1e6,zdir='z', offset=0,cmap=cm.cm.tempo,norm = norm,
            levels=clevs2)
cb_ax = fig.add_axes([0.80, 0.35, 0.01, 0.3])
cbar = fig.colorbar(cset, cax=cb_ax, cmap = cm.cm.tempo)
cbar.ax.set_ylabel('EKE (MJm$^{-2}$)', fontsize = 20)
ax.contourf(Xt, Yt,mask_array.values,zdir='z', offset=0,colors = 'darkgrey',
            levels=[1,2])

ax.contourf(Xss, mask_array3.values,-Zss/400,zdir='y', offset=-81,colors = 'lightgrey',
            levels=[1,2])

ax._axis3don = False

clevs2 = np.linspace(1032.5,1037.5,11)

cset = ax.contourf(rho0_front.values,Yf,-Zf/400, zdir='x',cmap=newcmap, offset =lonE,
             levels=clevs2,rstride=stride_value, cstride=stride_value)
ax.contourf(mask_array2.values,Yf,-Zf/400, zdir='x',colors = 'dimgrey', offset =lonE,
             levels=[1,2],rstride=stride_value, cstride=stride_value)
cb_ax = fig.add_axes([0.72, 0.35, 0.01, 0.3])
cbar = fig.colorbar(cset, cax=cb_ax, cmap = newcmap)
cbar.ax.set_ylabel('density (kgm$^{-3}$)',fontsize = 20)# # add front face of topo:
surf = ax.contour(Xt_land, Yt_land, np.nan_to_num(land_mask), [0],colors='0',rstride=stride_value, 
                       cstride=stride_value,linewidths=.9, zorder=1)
ax.view_init(20, -15)
ax.set_xlim(-173,-50)
ax.set_zlim(bottom=-14,top=0)

#fig = plt.savefig('RP_diagram_v2.png',dpi = 400)
```
